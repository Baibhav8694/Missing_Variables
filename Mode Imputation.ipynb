```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day37-handling-missing-categorical-data/train.csv', usecols= ['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

Most frequent value imputation - For categorical Data, the most repeated variable is replaced in the missing term

Assumptions while doing this - 
        1. Data Must be missing completely at random
        2. the value should repeat significant number of times to replace it

```python
df.isnull().mean()
```

```python
#for garage quality
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
plt.plot()
```

```python
#comparing the most frequent value with the other values
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual']== 'TA']['SalePrice'].plot(kind = 'kde', ax = ax)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind = 'kde', ax = ax, color = 'red')

lines, labels = ax.get_legend_handles_labels() 
labels = ['Houses with TA', 'Houses without TA']
ax.legend(lines, labels, loc = 'best')

plt.title('GrageQual')
plt.show()
```

```python
df['GarageQual'].mode()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde', ax=ax)

df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Houses with TA', 'Houses with NA']
ax.legend(lines, labels, loc='best')

plt.title('GarageQual')
plt.show()
```

```python
temp = df[df['GarageQual'] == 'TA']['SalePrice']
```

```python
df['GarageQual'].fillna('TA', inplace = True)
```

```python
df["GarageQual"].value_counts().plot(kind = 'bar')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

temp.plot(kind = 'kde', ax = ax)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind= 'kde', ax = ax, color = 'red')

labels = ['Original Variable', 'Imputed Variable']
ax.legend(lines, labels, loc = 'best')

#adding title
plt.title('GarageQuality')
plt.show()
```

since missing values are very less in garagequal  thus the graph overlaps and becomes almost equivalent to the original graph

```python
temp = df[df["FireplaceQu"]=='Gd']['SalePrice']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

temp.plot(kind = 'kde')

#distributionn after immputation
df[df["FireplaceQu"]=='Gd']['SalePrice'].plot(kind= 'kde', ax = ax, color = 'yellow')

lines, labels = ax.get_legend_handles_labels()
labels = ['Original variable', 'Imputed Variable']

ax.legend(lines, labels, loc= 'best')

#adding title
plt.title("FireplaceQu")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['SalePrice']), df['SalePrice'], test_size = 0.2)
```

```python
from sklearn.impute import SimpleImputer 

imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```
